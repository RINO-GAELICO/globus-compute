## Plotting results for Ionorb scaling with 4 processes and 10000 functions

```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dir = "results_ionorb/throughput/"

data = []
for i in range(1, 9):  
    filename = f"{i}_node_throughput_ionorb_1000_mps-endpoint_4_proc.json"
    full_path = dir + filename
    with open(full_path, 'r') as file:
        content = json.load(file)
        throughput = content["0"]["throughput"]
        throughput = 60 * throughput  # convert throughput to functions per minute
        data.append({"nodes": i, "throughput": throughput})

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data using Seaborn for median throughput
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x="nodes", y="throughput", data=df, marker="o", label="Median Throughput mps-endpoint with 8 processes")

# Calculate linear scaling
throughput_node_1 = df[df.nodes == 1].throughput.values[0]
linear_scaling = df.nodes * throughput_node_1

# Plot linear scaling using plt.plot()
plt.plot(df.nodes, linear_scaling, linestyle="--", color="red", label="Linear Scaling")

# Add labels for individual points using plt.text()
for i in range(len(df)):
    plt.text(df.nodes[i], df.throughput[i], f"{df.throughput[i]:.2f}", ha='right')

plt.title("Throughput vs Number of Nodes")
plt.xlabel("Number of Nodes")
plt.ylabel("Throughput (functions per minute)")
plt.legend()
plt.show()
```

## Plotting the results for one node augmenting the number of processes

```python
import json
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

file_dir = "results_ionorb/throughput/"
# Each file is named like 1_node_throughput_ionorb_4_mps-endpoint_4_proc
files = os.listdir(file_dir)
# only include files that have the correct format: 1_node_throughput_ionorb_X_mps-endpoint_X_proc
files = [file for file in files if file.endswith('.json') and '1_node_throughput_ionorb' in file and 'mps-endpoint' in file]
# the second number in the filename should be less than or equal to 64
files = [file for file in files if int(file.split('_')[4]) <= 64]
print("Length of files:", len(files))

configuration = 'mps-endpoint'
data_list = []

print(f"{configuration} files:", files)

for file in files:
    with open(os.path.join(file_dir, file), 'r') as f:
        data = json.load(f)
        # Extract the number of processes from the filename
        processes = int(file.split('_')[-2])
        throughputs = [data[key]['throughput'] for key in data]
        median_throughput = np.median(throughputs)
        q75, q25 = np.percentile(throughputs, [75, 25])
        iqr = q75 - q25
        
        data_list.append({
            'Processes': processes,
            'Median Throughput': median_throughput,
            'IQR': iqr
        })

# Convert data_list to DataFrame
df = pd.DataFrame(data_list)

# Sort the DataFrame by the number of processes
df = df.sort_values(by='Processes')

# Create a Seaborn plot
plt.figure(figsize=(15, 10))
sns.set_theme(style="darkgrid")

# Plot the median throughput
sns.lineplot(data=df, x='Processes', y='Median Throughput', marker='o', label=f'{configuration} - Median Throughput', color='m')

# Add the fill between for IQR
plt.fill_between(df['Processes'], 
                 df['Median Throughput'] - df['IQR'] / 2, 
                 df['Median Throughput'] + df['IQR'] / 2, 
                 color='m', alpha=0.2)

# Add labels for the markers
for i in range(len(df)):
    plt.text(df['Processes'].iloc[i], df['Median Throughput'].iloc[i] + 0.01, 
             f'{df["Median Throughput"].iloc[i]:.2f}', 
             horizontalalignment='center', color='black')

plt.xlabel('Number of Processes')
plt.ylabel('Throughput (functions per second)')
plt.title('Median Throughput and IQR vs Number of Processes for One Node')
plt.xticks(np.arange(df['Processes'].min(), df['Processes'].max()+1, 4))
plt.yticks(np.arange(int(df['Median Throughput'].min()), 
                     int(df['Median Throughput'].max())+1, 0.1))
plt.legend()
plt.show()
```
